### Gradient Decent를 이용한 선형회귀 구현

```python
import numpy as np
```

```python
# 배열을 만든다
# 이차원, 삼차원 배열인 (다차원배열) X는 대문자로 쓴다 (지금은 일차원)
# float32 type으로 만들어준다 (실수)
X = np.array([1, 2, 3], dtype = 'float32')
X
```

```python
# 배열을 만든다
# float32 type으로 만들어준다 (실수)
y = np.array([2, 2.5, 3.5], dtype = 'float32')
y
```

```python
# 초기 선형회귀식은 wX + b = 2X + 1 이라고 가정
w = 2
b = 1
```

```python
# hypothesis는 예측값
hypothesis = X * w + b
hypothesis
```


##### cost 함수는 제곱을 했기 때문에 곡선(사진모양)이 나온다


###### cost 에 1/2를 곱하는 이유는 어차피 0이 되면 0이기 때문에 계산의 편의를 위해 곱해준다

```python
# 빼기는 같은 자리에서 계산
hypothesis - y
```

```python
X
```

배열끼리 곱하기 할 때
np.dot 사용

```python
# 배열끼리 곱하기는 그냥 곱하기하면 안된다
# X, y는 배열
# np.dot 는 배열끼리 곱하기하는 방법임
# tensorflow 는 제대로 안나오고 줄, 칸을 맞추어주어야한다
w_gred = 1/3*np.dot((hypothesis-y), X)
w_gred
```

```python
b_gred = 1/3 * np.sum(hypothesis - y)
b_gred
```

```python
w = w - w_gred
w
```

```python
# MSE = cost
```

```python
for i in range(100):
    hypothesis = X*w+b
    print("hypothesis: ", hypothesis)
    cost = np.sum((hypothesis-y)**2)
    print("cost: ", cost)
    w_gred = 1/len(X) * np.dot((hypothesis-y),X.T)
    print("w_gred: ", w_gred)
    b_gred = 1/len(X) * np.sum(hypothesis -y)
    w = w-w_gred
    b = b-b_gred
    print('w: ',w, 'b: ',b)
```

```python
w = 2
b = 1
learning_rate = 0.1
for i in range(1000):
    hypothesis = X*w+b
    print("hypothesis: ",hypothesis)
    MSE = np.sum((hypothesis - y)**2)
    print("MSE: ",MSE)
    w_gred = 1/3*np.dot((hypothesis-y), X.T)
    print("w_gred: ",w_gred)
    b_gred = 1/3*np.sum(hypothesis-y)
    print("b_gred: ",b_gred)
    w = w-(learning_rate*w_gred)
    print("w: ",w)
    b = b-(learning_rate*b_gred)
    print("b: ", b)
```

```python
import tensorflow as tf
tf.__version__
import numpy as np
```

```python
tf.enable_eager_execution()
```

```python
X = np.array([1, 2, 3], dtype = "float32")
X
```

```python
Y = np.array([2.5, 5, 3.5], dtype = "float32")
Y
```

```python
# X에 곱해지는 값 W를 선언, 초기값은 2
W = tf.Variable([2], dtype = "float32")
print(W)
# b에 더해지는 값 b를 선언, 초기값은 1
b = tf.Variable([1], dtype = "float32")
print(b)
```

```python
hypothesis = X*W+b
hypothesis
```

tensorflow에서 제곱하려면 square 함수를 사용한다<br>
tensorflow에서 평균을 구하려면 reduce_mean 함수를 사용한다

```python
hypothesis - Y
```

```python
# tensorflow에서 배열 안에 있는 값을 제곱하는 방법 // tf.square
tf.square(hypothesis - Y)
```

```python
# 배열 안에 있는 값의 평균 구하기 // tf.reduce_mean
tf.reduce_mean(tf.square(hypothesis - Y))
```

```python
learning_rate = 0.1
```

```python
for i in range(1000):
    # 기울기를 구해준다
    with tf.GradientTape() as tape:
        hypothesis = X * W + b
        print("hypothesis: ", hypothesis)
        cost = tf.reduce_mean(tf.square(hypothesis - Y))
        print("cost: ", cost)
        
    w_grad, b_grad = tape.gradient(cost, [W, b])
    print("w_grad: ", w_grad, "b_grad: ", b_grad)
    # 그만큼 빼준다
    W.assign_sub(w_grad*learning_rate)
    b.assign_sub(b_grad*learning_rate)
    print("W: ", W)
    print("b: ", b)
    print("="*50)
```
