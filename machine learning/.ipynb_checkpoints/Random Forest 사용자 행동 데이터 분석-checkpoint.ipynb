```python
import pandas as pd
```

```python
# 데이터의 분류방법을 설정 : sep
# 각 데이터 파일들은 공백으로 분리되어 있으므로 read_csv에서 공백문자를 sep으로 할당
# \s+ : 1개 이상의 공백

feature_name_df = pd.read_csv('./human_activity_features.txt',
                             sep = '\s+',
                             header = None,
                             names = ['column_index','column_name'])
feature_name_df
```

```python
# feature_name_df의 [:, 1] 모든 줄 1번째 칸(컬럼이름) 조회
feature_name_df.iloc[:, 1]
```

```python
# feature_name_df의 [:, 1] 모든 줄 1번쩨 칸(컬럼이름)을 numpy() 배열로 변환해서 조회
feature_name_df.iloc[:,1].values
```

```python
# DataFrame에 피처명을 컬럼으로 부여하기 위해 numpy() 배열을 리스트 객체로 다시 변환
# numpy 배열을 list로 바꿀 수 있다 : tolist() 함수를 사용하면 가능
feature_name = feature_name_df.iloc[:,1].values.tolist()
feature_name
```

```python
# 학습데이터를 읽어서 DataFrame으로 만들어서 return
# sep: 데이터 분류 방법을 설정
# \s+ : 1개 이상의 공백
# header = None : 데이터 파일에 컬럼제목이 없음

X_train = pd.read_csv('./human_activity_X_train.txt', sep = '\s+', header = None)
X_train
```

```python
# X_train의 컬럼명으로 feature_name 대입
X_train.columns = feature_name
X_train
```

```python
# 각 컬럼의 통계 수치 조회
X_train.describe()
```

```python
# 테스트데이터를 읽어서 DataFrame으로 만들어서 return
# sep: 데이터의 분류방법을 설정
# \s+ : 1개 이상의 공백
# header = None : 데이터 파일에 컬럼 제목이 없음
X_test = pd.read_csv('./human_activity_X_test.txt', sep = '\s+', header = None)
X_test
```

```python
X_test.columns = feature_name
X_test
```

```python
# 동작 정보데이터를 DataFrame에 저장

# names = ['action'] : 컬럼명을 action으로 설정

# 1 WALKING
# 2 WALKINT_UPSTAIRS
# 3 WALKING_DOWNSTAIRS
# 4 SITTING
# 5 STANDING
# 6 LAYING

y_train = pd.read_csv('./human_activity_y_train.txt', sep = '\s+', header = None, names = ['action'])
y_train
```

```python
# 동작 정보데이터를 DataFrame에 저장

# names = ['action'] : 컬럼명을 action으로 설정

# 1 WALKING
# 2 WALKINT_UPSTAIRS
# 3 WALKING_DOWNSTAIRS
# 4 SITTING
# 5 STANDING
# 6 LAYING

y_test = pd.read_csv('./human_activity_y_test.txt', sep = '\s+', header = None, names = ['action'])
y_test
```

```python
# 학습데이터의 각 항목별 빈도수를 조회
# 1 WALKING
# 2 WALKINT_UPSTAIRS
# 3 WALKING_DOWNSTAIRS
# 4 SITTING
# 5 STANDING
# 6 LAYING

y_train['action'].value_counts()
```

```python
# 학습데이터의 null 값의 합을 조회
X_train.isnull().sum()
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
import pandas as pd
import warnings
import numpy as np
```

```python
# RandomForestClassifier 객체 생성해서 rf_clf에 대입
rf_clf = RandomForestClassifier()
rf_clf
```

```python
# 학습데이터 X_train과 y_train을 이용하여 트리플 생성
rf_clf.fit(X_train, y_train)
```

```python
# rf_clf.predict(X_test): X_test데이터를 예측하고 결과를 RETURN
pred = rf_clf.predict(X_test)
pred
```

```python
# accuracy_score(y_test, pred): y_test와 pred를 비교해서 정확도를 출력
accuracy = accuracy_score(y_test, pred)
print("랜덤 포레스트 정확도: {}".format(accuracy))
```

```python
from sklearn.model_selection import GridSearchCV
# Pandas Forest의 성능 향상을 위해서 옵션을 설정
params = {
    'n_estimators': [100], #트리의 개수 100개
    'max_depth': [6, 8, 10, 12], # 최대 트리의 깊이
    'min_samples_leaf': [8, 12, 18], # 최소 트리 잎의 개수
    'min_samples_split': [8, 16, 20] # 전체 데이터의 로우 중에서
                                     # 최소 몇개의 데이터를 이용해서 트리를 만들 것인가
    }
```

```python
# RandomForestClassifier 객체 생성
# n_jobs = 10: CPU를 10개로 분할해서 동시에 10개씩 테스트

rf_clf = RandomForestClassifier(n_jobs = 10)
rf_clf
```

```python
# GridSearchCV 객체를 이용해서  RandomForestClassifier 객체의 최적의 파라미터를 찾음
# GridSerachCV(최적의 파라미터를 찾을 RandomForest 객체, param_grid = 파라미터 옵션이 저장된 객체)
# n_jobs = 10 : CPU를 10개로 분할해서 동시에 10개씩 테스트

grid_cv = GridSearchCV(rf_clf, param_grid = params, n_jobs = 10)
grid_cv
```

```python
warnings.filterwarnings('ignore')
```

```python
# grid_cv.fit(X_train, y_train): X_train, y_train을 이용하여 트리들을 만들고
# 설정된 파라미터의 모든 경우 수의 정확도를 테스트해서 가장 정확도가 높은 파라미터를 찾음
grid_cv.fit(X_train, y_train)
```

```python
print('최적 하이퍼 파라미터: \n', grid_cv.best_params_)
print('최고 예측 정확도: {0:.4f}'.format(grid_cv.best_score_))
```

Random Forest 컬럼의 중요도

```python
rf_clf1 = RandomForestClassifier(n_estimators = 300,
                                max_depth = 10,
                                min_samples_leaf = 8,
                                min_samples_split = 8)
rf_clf1
```

```python
rf_clf1.fit(X_train, y_train)
```

```python
pred = rf_clf1.predict(X_test)
pred
```

```python
print('예측 정확도: {}'.format(accuracy_score(y_test, pred)))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
ftr_importances_values = rf_clf1.feature_importances_
ftr_importances_values
```

```python
ftr_importances = pd.Series(ftr_importances_values, index = X_train.columns)
ftr_importances
```

```python
ftr_top20 = ftr_importances.sort_values(ascending = False)[:20]
ftr_top20
```

```python
plt.figure(figsize=(8,6))
plt.title('Feature importance Top 20')
sns.barplot(x=ftr_top20, y=ftr_top20.index)
plt.show()
```
