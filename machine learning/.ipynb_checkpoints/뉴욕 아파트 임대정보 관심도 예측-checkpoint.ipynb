```python
# df['photos'].apply(len): photos 컬럼의 각 줄의 데이터에 len 함수를 실행해서
# 사진의 개수를 리턴하고 데이터를 df['num_photos']에 저장
df['num_photos'] = df['photos'].apply(len)
df['num_photos'].head()
```

```python
# df['features'].apply(len): photos 컬럼의 각 줄의 데이터에 len 함수를 실행해서 특정의
# 개수를 리턴하고 데이터를 df['num_feature']에 저장
df['num_features'] = df['features'].apply(len)
df['num_features'].head()
```

```python
# df['description'].apply(lambda x: len(x,split(' '))): description 컬럼의 데이터를 x에 대입
#                                  x를 공백을 기준으로 단어를 분리 (x,split(' '))
#                                  하고 단어의 개수를 리턴 len(x,split(' ')))
# 데이터를 df['num_words'] 컬럼에 저장
df['num_words'] = df['description'].apply(lambda x: len(x.split(' ')))
df['num_words'].head()
```

```python
# interest_level이 문자여서 에러 발생 // 숫자형태로 나올 수 있도록 interest 컬럼 생성

# np.where : 조건에 따라서 다른 값을 리턴하는 함수
# df['interest_level']=='low',0 : interest_level 컬럼의 값이 low이면 0 리턴
# df['interest_level']=='medium',1,2 : interest_level 컬럼의 값이 
#                                      medium이면 1, 아니면 2리턴

df['interest'] = np.where(df['interest_level']=='low',0,
                          np.where(df['interest_level']=='medium',1,2))
```

```python
df[['interest_level','interest']].head()
```

```python
features_to_use01=['bathrooms','bedrooms','latitude','longitude',
                  'price','year','month','day','week','hour','weekday',
                  'num_photos','num_features','num_words']
```

```python
X = df[features_to_use01]
X.head()
```

```python
import missingno as msno
# msno.bar는 각 컬럼의 mull 값이 아닌 데이터의 비율을 그래프로 그려주는 패키지
msno.bar(df=X)
```

X에 결측치는 없다

```python
X = np.array(X,dtype="float32")
X
```

```python
y = df['interest']
y.head()
```

```python
y = np.array(y, dtype="float32")
y
```

```python
y = y.reshape(-1, 1)
y
```

```python
from sklearn.model_selection import train_test_split
# 학습데이터와 테스트 데이터 분리
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 1234)
```

```python
X_train.shape
```

```python
X_test
```

```python
y_train
```

```python
y_test
```

```python
from sklearn.preprocessing import OneHotEncoder
enc = OneHotEncoder()
enc.fit(y)
```

```python
y_train_onehot = enc.transform(y_train).toarray()
y_train_onehot
```

```python
y_test_onehot = enc.transform(y_test).toarray()
y_test_onehot
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X_train)
```

```python
X_train = scaler.transform(X_train)
X_train
```

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import Adam
```

```python
model = Sequential()
model.add(Dense(200,input_dim=14,activation="sigmoid"))
```

```python
model.add(Dense(100, activation="sigmoid"))
```

```python
model.add(Dense(3, activation="softmax"))
```

```python
model.summary()
```

```python
model.compile(loss="categorical_crossentropy", optimizer=Adam(lr=0.01),metrics=['acc'])
```

```python
model.fit(X_train, y_train_onehot, epochs=100)
```

```python
X_test = scaler.transform(X_test)
X_test
```

```python
predict = model.predict(X_test)
predict
```

```python
predict01 = np.argmax(predict, 1)
predict01
```

```python
target = np.argmax(y_test_onehot, 1)
target
```

```python
predict02 = (predict01 == target)
predict02
```

```python
np.sum(predict02)
```

```python
acc = np.sum(predict02) / len(predict02)
acc
```
