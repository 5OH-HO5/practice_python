```python
from sklearn import datasets
from sklearn.model_selection import train_test_split
import numpy as np
import tensorflow as tf
tf.enable_eager_execution()
# w    tf.enable_eager_execution()   
```

```python
# iris data 
data = datasets.load_iris()
data
```

```python
# IRIS    X 
X = data.data
X
# : cm //   /   /   /  
```

```python
X = np.array(X, dtype="float32")
X
```

```python
# IRIS 
# setosa: 0
# versicolor: 1
# virginica: 2
y = data.target
y
```

```python
# y.reshape(): 1 2 
y = y.reshape(150, 1)
# y.reshape(-1, 1)  // -1    
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
X_train
```

```python
X_test
```

```python
y_train
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
#  one hot   
enc = OneHotEncoder()
# y   
enc.fit(y)
```

```python
# y_train one hot encoding
y_train_onehot = enc.transform(y_train).toarray()
y_train_onehot
```

```python
#  float64  tensorflow float32  
# y_train_onehot float32 type 
y_train_onehot = np.array(y_train_onehot, dtype="float32")
y_train_onehot
```

```python
# y_test  onehotencoding 
y_test_onehot = enc.transform(y_test).toarray()
y_test_onehot
```

```python
y_test_onehot = np.array(y_test_onehot, dtype="float32")
y_test_onehot
```

```python
from sklearn.preprocessing import StandardScaler
# scaler     0 ~ 1 
scaler = StandardScaler()
# scaler.fit(X_train): X_train ,  
scaler.fit(X_train)
```

```python
# scaler.transtorm(X_train): X_train     X_train 
X_train = scaler.transform(X_train)
X_train
```

```python
# tf.contrib.layers.xavier_initializer(): w b    
initializer = tf.contrib.layers.xavier_initializer()
```

```python
# tf.Variable(initializer([4,16])): 4 16 w0  return
# w0 = initializer([4,16])
# tensor type     
# cost 0   tensorflow  Variable 
w0 = tf.Variable(initializer([4,16]))
w0
```

```python
# tf.Variable(initializer([16])): 16 1 b0  return
b0 = tf.Variable(initializer([16]))
b0
```

```python
tf.matmul(X_train, w0) + b0
```

```python
hypothesis0 = tf.sigmoid(tf.matmul(X_train, w0) + b0)
hypothesis0
```

```python
w1 = tf.Variable(initializer([16,3]))
w1
```

```python
b1 = tf.Variable(initializer([3]))
b1
```

```python
hypothesis1 = tf.nn.softmax(tf.matmul(hypothesis0, w1) + b1)
hypothesis1
```

```python
# w0, w1, w2 b0, b1, b2    Optimizer  
optimizer = tf.train.AdamOptimizer(0.1)
```

```python
#  : tape
# w b : AdamOptimizer

for step in range(1000):
    with tf.GradientTape() as tape:
        hypothesis0 = tf.sigmoid(tf.matmul(X_train, w0) + b0)
        hypothesis1 = tf.nn.softmax(tf.matmul(hypothesis0, w1) + b1)
        
        cost = -tf.reduce_sum(y_train_onehot * tf.log(hypothesis1)) / len(hypothesis1)
        
        grads = tape.gradient(cost, [w0,w1,b0,b1])
        
        optimizer.apply_gradients(grads_and_vars = zip(grads, [w0,w1,b0,b1]))
        
        if step % 100 == 0:
            print("="*50)
            print("step: {}, cost: {}".format(step, cost.numpy()))
            print("="*50)
```

```python
# scaler  X_test  
#  scaler,fit(X_train)    scaler X_train ,   
# scaler.transform(X_test): X_test X_train   X_train  
X_test = scaler.transform(X_test)
X_test
```

```python
hypothesis0 = tf.sigmoid(tf.matmul(X_test, w0) + b0)
predict = tf.nn.softmax(tf.matmul(hypothesis0, w1) + b1)
predict
```

```python
# tf.argmax(predict, 1): predict   return
#                    1:     return
predict01 = tf.argmax(predict, 1)
predict01
```

```python
# tf.argmax(y, 1): y   return
#              1:     return
target = tf.argmax(y_test_onehot, 1)
target
```

```python
ac01 = tf.equal(predict01, target)
ac01
```

```python
ac02 = tf.cast(ac01, dtype="float32")
ac02
```

```python
ac03 = tf.reduce_mean(ac02)
ac03
```

```python
#  
#   

accuracy = tf.reduce_mean( tf.cast(tf.equal(predict01, target), dtype = tf.float32))
print("="*50)
print("accuracy")
print(accuracy.numpy())
```
