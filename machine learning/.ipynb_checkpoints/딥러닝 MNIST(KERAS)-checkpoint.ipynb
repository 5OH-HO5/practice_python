```python
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import Adam
import numpy as np
```

```python
from keras.datasets import mnist
# mnist, load_data(): 케라스에 저장된 MNIST 이미지를 return
# (X_train, y_train), (X_test, y_test) 순으로 리턴받아서 변수에 저장
(X_train, y_train), (X_test, y_test) = mnist.load_data()
```

```python
X_train
```

```python
# X_train의 shape 조회 -> 3차원 배열
X_train.shape
```

```python
# X_train을 2차원 배열로 변환
# X_train.reshape(새로운 줄, 새로운 칸)

# 줄: 60000, 칸: 28 * 28 = 784
X_train = X_train.reshape(60000, 784)

# X_train의 줄, 칸 조회
X_train.shape
```

```python
X_train
```

```python
y_train
```

```python
# y_train.reshape(): 1차원 배열을 2차원 배열로 변환
# y_train.reshape(1, -1): 행 형태의 2차원 배열로 변환
#                       [[1,0,1,0,1,0,1,0 ...]]
# y_train.reshape(-1, 1): 열 형태의 2차원 배열로 변환
#                       [
#                       [0],
#                       [1],
#                       [0],
#                       ,,,]

y_train = y_train.reshape(-1,1)
y_train
```

```python
# X_test의 shape 조회 -> 3차원 배열
X_test.shape
```

```python
# X_test을 2차원 배열로 변환
# X_test.reshape(새로운 줄, 새로운 칸)

X_test = X_test.reshape(10000,784)

X_test.shape
```

```python
X_test
```

```python
y_test
```

```python
# 1차원 배열을 2차원 배열로 변환
y_test = y_test.reshape(-1, 1)
y_test
```

```python
print("학습셋 이미지 수: {}개".format(X_train.shape[0]))
print("테스트셋 이미지 수: {}개".format(X_test.shape[0]))
```

```python
from sklearn.preprocessing import OneHotEncoder
# 데이터를 onehot인코딩 할 객체
enc = OneHotEncoder()
# y_train가 몇가지 종류인지를 파악
enc.fit(y_train)
```

```python
# enc.transform(y_train): y_train을 OneHotEncoding
# toarray(): OneHotEncording 한 데이터를 numpy 배열로 변환
y_train_onehot = enc.transform(y_train).toarray()
y_train_onehot
```

```python
# y_train_onehot을 float32 타입으로 변환
y_train_onehot = np.array(y_train_onehot, dtype = "float32")
y_train_onehot
```

```python
# enc.transform(y_test): y_test을 OneHotEncoding
# toarray(): OneHotEncording 한 데이터를 numpy 배열로 변환
y_test_onehot = enc.transform(y_test).toarray()
y_test_onehot
```

```python
# y_test_onehot을 float32 타입으로 변환
y_test_onehot = np.array(y_test_onehot, dtype = "float32")
y_test_onehot
```

```python
# 입력값을 읽어서 예측을 할 Sequential 객체 생성
model = Sequential()

model.add(Dense(200, input_dim=784, activation="sigmoid"))
```

```python
# Dense(출력데이터의 칸의 수): 선형회귀를 수행할 객체

# 두번쨰부터는 'input_dim' (여기서는 input_dim = 200) 생략가능
model.add(Dense(10, activation="softmax"))
```

```python
# param counting 하는 방법 궁금
model.summary()
```

```python
model.compile(loss="categorical_crossentropy",optimizer=Adam(lr=0.001),metrics=['acc'])
```

```python
model.fit(X_train,y_train_onehot,epochs=10,batch_size=100)
```

```python
X_test
```

```python
# X_test가 어떤 이미지인지 분류
predict = model.predict(X_test)
predict
```

```python
predict01 = np.argmax(predict, 1)
predict01
```

```python
y_test_onehot
```

```python
target = np.argmax(y_test_onehot, 1)
target
```

```python
predict02 = (predict01 == target)
predict02
```

```python
np.sum(predict02)
```

```python
acc = np.sum(predict02) / len(predict02)
acc
```

```python
# 예상치와 실제 이미지를 출력하기 위한 패키지
import matplotlib.pyplot as plt
```

```python
# 이미지를 출력할 객체
fig = plt.figure()
for i in range(10):
    # 2줄 5칸으로 나누고 i + 1번째 이미지를 그림
    subplot = fig.add_subplot(2, 5, i + 1)
    
    # subplot.set_xticks([]) : 그래프의 x축을 설정
    # 데이터가 없으므로 x축에 아무 데이터도 출력 안됨
    subplot.set_xticks([])
    
    # subplot.set_yticks([]) : 그래프의 y축을 설정
    # 데이터가 없으므로 y축에 아무 데이터도 출력 안됨
    subplot.set_yticks([])
    
    # 이미지의 제목으로 예측한 이미지의 이름 출력
    subplot.set_title("predict: %d" % predict01[i])
    # 이미지 출력
    subplot.imshow(X_test[i].reshape(28, 28), cmap = plt.cm.gray_r)
# 그래프를 화면에 그림
plt.show()
```
