```python
import numpy as np
import matplotlib.pyplot as plt
from keras.preprocessing.image import ImageDataGenerator
```

```python
# 데이터 경로
TRAIN_DIR = './dental_image/train/'
```

```python
# 이미지의 가로 세로
IM_WIDTH = 220
IM_HEIGHT = 220
```

```python
# 학습에 사용할 이미지를 읽을 객체
datagen = ImageDataGenerator()
```

```python
# datagen.flow_from_directory(): 디렉토리에서 이미지를 읽을 객체 생성
# datagen.flow_from_directory(
#                            이미지 경로
#                            target_size: 이미지의 가로세로
#                            batch_size: 한번에 리턴할 이미지 개수
#                            class_mode: 분류방법 / categorical: (다중분류)
#                                                   binary(0, 1로 이중분류)
#                            shuffle: 이미지를 섞을 여부
train_generator = datagen.flow_from_directory(
    TRAIN_DIR,
    target_size=(IM_WIDTH, IM_HEIGHT),
    batch_size=30,
    class_mode='categorical',
    shuffle=True)
```

```python
# 학습데이터의 전체 이미지 개수를 리턴
train_generator.n
```

```python
# 한번에 리턴할 이미지의 개수
train_generator.batch_size
```

```python
# train_generator.next()
# train_generator.batch_size에 설정된 이미지 30개를 리턴해서 img에 대입
# 이미지의 label을 label에 대입
img, label = train_generator.next()
```

```python
# one_hot encoding이 되어있는 것 확인
train_generator.class_indices
```

```python
# np.argmax(label,1): label의 최대값의 인덱스를 리턴
#                 1: 각 줄에서 최대값의 인덱스를 리턴
label_num = np.argmax(label,1)
label_num
```

```python
type(train_generator.class_indices)
```

```python
# keys(): 글씨만 리턴
label_key = train_generator.class_indices.keys()
label_key
```

```python
label_key = list(train_generator.class_indices.keys())
label_key
```

```python
label_key[label_num[0]]
```

```python
# label_num에서 하나를 꺼내서 num에다가 넣는다
# 그 num을 label_key에 넣으면 label_str에 저장
label_str = [label_key[num] for num in label_num]
label_str
```

```python
# 숫자는 %d / 문자는 %s
plt.title("label: %s" % (label_str[0]))
plt.imshow((img[0] / 255.0))
```

```python
# 이미지를 출력할 객체
fig = plt.figure(figsize=(30, 20))

for i in range(30):
    # 6줄 5칸(총 30)으로 나누고 i+1번째 이미지를 그림
    subplot = fig.add_subplot(6, 5, i + 1)
    
    # 데이터가 없으므로 x축에 아무 데이터도 출력 안함
    subplot.set_xticks([])
    
    # 데이터가 없으므로 y축에 아무 데이터도 출력 안함
    subplot.set_yticks([])
    
    # 이미지의 제목으로 예측한 이미지의 이름 출력
    # 숫자는 %d / 문자는 %s
    plt.title("label: %s" % (label_str[i]))
    
    # img[i]에 저장된 이미지를 255로 나눠서 0~1 사이의 데이터로 변환
    img[i] = img[i] / 255.0
    
    # 이미지 출력
    subplot.imshow(img[i])

# 그래프를 화면에 그림
plt.show()
```
