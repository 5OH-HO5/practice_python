```python
import pandas as pd
```

```python
# 데이터의 분류방법을 설정 : sep
# 각 데이터 파일들은 공백으로 분리되어 있으므로 read_csv에서 공백문자를 sep으로 할당
# \s+ : 1개 이상의 공백

feature_name_df = pd.read_csv('./human_activity_features.txt',
                             sep = '\s+',
                             header = None,
                             names = ['column_index','column_name'])
feature_name_df
```

```python
# feature_name_df의 [:, 1] 모든 줄 1번째 칸(컬럼이름) 조회
feature_name_df.iloc[:, 1]
```

```python
# feature_name_df의 [:, 1] 모든 줄 1번쩨 칸(컬럼이름)을 numpy() 배열로 변환해서 조회
feature_name_df.iloc[:,1].values
```

```python
# DataFrame에 피처명을 컬럼으로 부여하기 위해 numpy() 배열을 리스트 객체로 다시 변환
# numpy 배열을 list로 바꿀 수 있다 : tolist() 함수를 사용하면 가능
feature_name = feature_name_df.iloc[:,1].values.tolist()
feature_name
```

```python
# 학습데이터를 읽어서 DataFrame으로 만들어서 return
# sep: 데이터 분류 방법을 설정
# \s+ : 1개 이상의 공백
# header = None : 데이터 파일에 컬럼제목이 없음

X_train = pd.read_csv('./human_activity_X_train.txt', sep = '\s+', header = None)
X_train
```

```python
# X_train의 컬럼명으로 feature_name 대입
X_train.columns = feature_name
X_train
```

```python
# 각 컬럼의 통계 수치 조회
X_train.describe()
```

```python
# 테스트데이터를 읽어서 DataFrame으로 만들어서 return
# sep: 데이터의 분류방법을 설정
# \s+ : 1개 이상의 공백
# header = None : 데이터 파일에 컬럼 제목이 없음
X_test = pd.read_csv('./human_activity_X_test.txt', sep = '\s+', header = None)
X_test
```

```python
X_test.columns = feature_name
X_test
```

```python
# 동작 정보데이터를 DataFrame에 저장

# names = ['action'] : 컬럼명을 action으로 설정

# 1 WALKING
# 2 WALKINT_UPSTAIRS
# 3 WALKING_DOWNSTAIRS
# 4 SITTING
# 5 STANDING
# 6 LAYING

y_train = pd.read_csv('./human_activity_y_train.txt', sep = '\s+', header = None, names = ['action'])
y_train
```

```python
# 동작 정보데이터를 DataFrame에 저장

# names = ['action'] : 컬럼명을 action으로 설정

# 1 WALKING
# 2 WALKINT_UPSTAIRS
# 3 WALKING_DOWNSTAIRS
# 4 SITTING
# 5 STANDING
# 6 LAYING

y_test = pd.read_csv('./human_activity_y_test.txt', sep = '\s+', header = None, names = ['action'])
y_test
```

```python
# 학습데이터의 각 항목별 빈도수를 조회
# 1 WALKING
# 2 WALKINT_UPSTAIRS
# 3 WALKING_DOWNSTAIRS
# 4 SITTING
# 5 STANDING
# 6 LAYING

y_train['action'].value_counts()
```

```python
# 학습데이터의 null 값의 합을 조회
X_train.isnull().sum()
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
import pandas as pd
import warnings
import numpy as np
```

```python
# RandomForestClassifier 객체 생성해서 rf_clf에 대입
rf_clf = RandomForestClassifier()
rf_clf
```

```python
# 학습데이터 X_train과 y_train을 이용하여 트리플 생성
rf_clf.fit(X_train, y_train)
```

```python
# rf_clf.predict(X_test): X_test데이터를 예측하고 결과를 RETURN
pred = rf_clf.predict(X_test)
pred
```

```python
# accuracy_score(y_test, pred): y_test와 pred를 비교해서 정확도를 출력
accuracy = accuracy_score(y_test, pred)
print("랜덤 포레스트 정확도: {}".format(accuracy))
```
