```python
df['interest'] = np.where(df['interest_level'] == 'low', 0,
                         np.where(df['interest_level'] == 'medium',1,2))
```

```python
df[['interest_level','interest']].head()
```

### New Col

```python
df['price'] = df['price'].clip(upper = 13000)
```

```python
df['logprice'] = np.log(df['price']+1e-6)
df['logprice'] = np.array(df['logprice'], dtype = 'float32')
```

```python
df['roomsum'] = (df['bathrooms'] + df['bedrooms'])+1e-6
df['roomsum'] = df['roomsum'].astype('float32')
```

```python
df['PricePerRoom'] = (df['price'] / df['roomsum'])+1e-6
df['PricePerRoom'] = df['PricePerRoom'].astype('float32')
```

```python
df['logPPR'] = np.log(df['PricePerRoom']+1e-6)
df['logPPR'] = np.array(df['logPPR'], dtype = 'float32')
```

```python
df['PricePerBed'] = (df['price'] / (df['bedrooms']+1))+1e-6
df['PricePerBed'] = df['PricePerBed'].astype('float32')
```

```python
df['logPPB'] = np.log(df['PricePerBed']+1e-6)
df['logPPB'] = np.array(df['logPPB'], dtype = 'float32')
```

```python
df['bathrooms'] = df['bathrooms'].astype('float32')
df['latitude'] = df['latitude'].astype('float32')
df['longitude'] = df['longitude'].astype('float32')
```

```python
features_to_use01 = ['bathrooms','bedrooms','latitude','longitude','price',
                    'year','month','day','week','hour','weekday','num_photos',
                    'num_features','num_words','interest','logprice','roomsum',
                    'PricePerRoom','logPPR','PricePerBed','logPPB']
```

### Over Sampling

```python
X = df[features_to_use01]
X.head()
```

```python
y = X.interest
y
```

```python
Copy_X = X.copy().drop(['interest'], axis=1)
Copy_X
```

```python
smote_nc = SMOTENC(categorical_features=[0,3], random_state=0)
```

```python
X_sampled, y_sampled = smote_nc.fit_resample(Copy_X, y)
```

```python
sorted(Counter(y_sampled).items())
```

```python
X = X_sampled
y = y_sampled
```

```python
np.histogram(y)
```

```python
X = np.array(X, dtype = 'float32')
```

```python
y = np.array(y, dtype = 'float32')
```

```python
y = y.reshape(-1,1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = .2, random_state = 1234)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
enc = OneHotEncoder()
enc.fit(y)
```

```python
ytrao = enc.transform(y_train).toarray()
ytrao
```

```python
yteso = enc.transform(y_test).toarray()
yteso
```

```python
scaler = StandardScaler()
scaler.fit(X_train)
```

```python
X_train = scaler.transform(X_train)
```

```python
model = Sequential()
model.add(Dense(200, input_dim=20, activation='relu'))
model.add(Dropout(0.1))
```

```python
model.add(Dense(100, activation='relu'))
model.add(Dropout(0.1))
```

```python
model.add(Dense(70, activation='relu'))
model.add(Dropout(0.1))
```

```python
model.add(Dense(50, activation='relu'))
model.add(Dropout(0.1))
```

```python
model.add(Dense(30, activation='relu'))
model.add(Dropout(0.1))
```

```python
model.add(Dense(10, activation='relu'))
model.add(Dropout(0.1))
```

```python
model.add(Dense(3, activation='softmax'))
```

```python
model.summary()
```

```python
model.compile(loss = 'categorical_crossentropy', optimizer=Adam(lr=.0001),metrics = ['acc'])
model.fit(X_train, ytrao, epochs=500)
```

```python
X_test = scaler.transform(X_test)
X_test
```

```python
predict = model.predict(X_test)
predict.sum(axis = 0)
```

```python
np.sum(yteso, axis=0)
```

```python
predict01 = np.argmax(predict,1)
target = np.argmax(yteso, 1)
```

```python
predict02 = (predict01 == target)
predict02
```

```python
acc = np.sum(predict02) / len(predict02)
acc
```
