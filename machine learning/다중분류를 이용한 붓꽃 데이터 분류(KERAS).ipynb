```python
from sklearn import datasets
from sklearn.model_selection import train_test_split
import numpy as np
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import Adam
```

```python
data = datasets.load_iris()
data
```

```python
X = data.data
X
```

```python
y = data.target
y
```

```python
y = y.reshape(-1, 1)
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
X_train
```

```python
X_test
```

```python
y_train
```

```python
y_test
```

```python
# X에서 평균을 빼고 표준편차로 나눠서 평균 -> 0, 표준편차 -> 1로 변환할 객체 생성
# scaler.fit(--): --의 평균, 표준편차를 계산
# scaler.transform(--): --에서 평균을 뺴고 표준편차로 나눈다
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X_train)
```

```python
X_train = scaler.transform(X_train)
X_train
```

```python
from sklearn.preprocessing import OneHotEncoder
# 데이터를 one hot 인코딩 할 객체
enc = OneHotEncoder()
# y가 몇가지 종류인지 파악
enc.fit(y)
```

```python
# toarray(): one hot encoding 한 데이터를 numpy 배열로 변환
y_train_onehot = enc.transform(y_train).toarray()
y_train_onehot
```

```python
y_train_onehot = np.array(y_train_onehot, dtype = "float32")
y_train_onehot
```

```python
y_test_onehot = enc.transform(y_test).toarray()
y_test_onehot
```

```python
y_test_onehot = np.array(y_test_onehot, dtype="float32")
y_test_onehot
```

```python
model = Sequential()
model.add(Dense(16, input_dim=4, activation="sigmoid"))
```

```python
model.add(Dense(3, activation = "softmax"))
```

```python
# param counting 하는 방법 궁금!!
model.summary()
```

```python
model.compile(loss="categorical_crossentropy", optimizer=Adam(lr=0.1),metrics=['acc'])
```

```python
model.fit(X_train, y_train_onehot, epochs=1000)
```

```python
X_test = scaler.transform(X_test)
X_test
```

```python
predict = model.predict(X_test)
predict
```

```python
# np.argmax(predict, 1): predict 최대값의 인덱스를 리턴
#                    1: 각 줄에서 최대값의 인덱스를 리턴
predict01 = np.argmax(predict, 1)
predict01
```

```python
target = np.argmax(y_test_onehot, 1)
target
```

```python
predict02 = (predict01 == target)
predict02
```

```python
np.sum(predict02)
```

```python
# 정확도를 계산
# len(predict02): predict02의 줄의 수를 리턴
acc = np.sum(predict02) / len(predict02)
acc
```
