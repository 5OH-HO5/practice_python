```python
from sklearn import datasets
from sklearn.model_selection import train_test_split
import numpy as np
import tensorflow as tf
tf.enable_eager_execution()
# w값 출력이 안되는 경우는 tf.enable_eager_execution() 이거 안해줘서 그렇다
```

```python
# iris data에 저장
data = datasets.load_iris()
data
```

```python
# IRIS 판단의 독립변수를 읽어서 X에 저장
X = data.data
X
# 단위: cm // 꽃받침 길이 / 꽃받침 넓이 / 꽃잎 길이 / 꽃잎 넓이
```

```python
X = np.array(X, dtype="float32")
X
```

```python
# IRIS 종속변수
# setosa: 0
# versicolor: 1
# virginica: 2
y = data.target
y
```

```python
# y.reshape(): 1차원배열을 2차원배열로 변환
y = y.reshape(150, 1)
# y.reshape(-1, 1)과 동일 // -1은 줄 또는 칸을 무시
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
y_train
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
# 데이터를 one hot 인코딩 할 객체
enc = OneHotEncoder()
# y가 몇가지 종류인지를 파악
enc.fit(y)
```

```python
# y_train을 one hot encoding
y_train_onehot = enc.transform(y_train).toarray()
y_train_onehot
```

```python
# 기본타입은 float64 이지만 tensorflow가 float32 타입을 지원한다
# y_train_onehot을 float32 type으로 변환
y_train_onehot = np.array(y_train_onehot, dtype="float32")
y_train_onehot
```

```python
# y_test 도 onehotencoding 함
y_test_onehot = enc.transform(y_test).toarray()
y_test_onehot
```

```python
y_test_onehot = np.array(y_test_onehot, dtype="float32")
y_test_onehot
```

```python
from sklearn.preprocessing import StandardScaler
# scaler를 사용해서 범위를 맞추어준다  0 ~ 1 사이
scaler = StandardScaler()
# scaler.fit(X_train): X_train의 평균, 표준편차를 계산
scaler.fit(X_train)
```

```python
# scaler.transtorm(X_train): X_train에서 평균을 빼고 표준편차로 나눠서 X_train에 대입
X_train = scaler.transform(X_train)
X_train
```

```python
# tf.contrib.layers.xavier_initializer(): w와 b를 초기화 시키는 객체를 생성
initializer = tf.contrib.layers.xavier_initializer()
```

```python
# tf.Variable(initializer([4,16])): 4줄 16칸의 w0를 생성해서 return
# w0 = initializer([4,16])
# tensor type은 상수이며 값을 바꿀 수 없다
# cost가 0이 되는 것을 tensorflow가 찾아줘야해서 Variable로 형변환
w0 = tf.Variable(initializer([4,16]))
w0
```

```python
# tf.Variable(initializer([16])): 16줄 1칸의 b0을 생성해서 return
b0 = tf.Variable(initializer([16]))
b0
```

```python
tf.matmul(X_train, w0) + b0
```

```python
hypothesis0 = tf.sigmoid(tf.matmul(X_train, w0) + b0)
hypothesis0
```

```python
w1 = tf.Variable(initializer([16,3]))
w1
```

```python
b1 = tf.Variable(initializer([3]))
b1
```

```python
hypothesis1 = tf.nn.softmax(tf.matmul(hypothesis0, w1) + b1)
hypothesis1
```

```python
# w0, w1, w2와 b0, b1, b2 값을 자동으로 수정하는 Optimizer 객체를 생성
optimizer = tf.train.AdamOptimizer(0.1)
```

```python
# 기울기 계산: tape
# w와 b 계산: AdamOptimizer

for step in range(1000):
    with tf.GradientTape() as tape:
        hypothesis0 = tf.sigmoid(tf.matmul(X_train, w0) + b0)
        hypothesis1 = tf.nn.softmax(tf.matmul(hypothesis0, w1) + b1)
        
        cost = -tf.reduce_sum(y_train_onehot * tf.log(hypothesis1)) / len(hypothesis1)
        
        grads = tape.gradient(cost, [w0,w1,b0,b1])
        
        optimizer.apply_gradients(grads_and_vars = zip(grads, [w0,w1,b0,b1]))
        
        if step % 100 == 0:
            print("="*50)
            print("step: {}, cost: {}".format(step, cost.numpy()))
            print("="*50)
```

```python
# scaler를 이용해서 X_test의 범위를 변경함
# 위에서 scaler,fit(X_train) 함수를 실행했기 때문에 scaler에 X_train의 평균, 표준편차가 저장되어 있음
# scaler.transform(X_test): X_test에서 X_train의 평균을 빼고 X_train의 표준편차로 나눠줌
X_test = scaler.transform(X_test)
X_test
```

```python
hypothesis0 = tf.sigmoid(tf.matmul(X_test, w0) + b0)
predict = tf.nn.softmax(tf.matmul(hypothesis0, w1) + b1)
predict
```

```python
# tf.argmax(predict, 1): predict 최대값의 인덱스를 return
#                    1: 각 줄에서 최대값의 인덱스를 return
predict01 = tf.argmax(predict, 1)
predict01
```

```python
# tf.argmax(y, 1): y 최대값의 인덱스를 return
#              1: 각 줄에서 최대값의 인덱스를 return
target = tf.argmax(y_test_onehot, 1)
target
```

```python
ac01 = tf.equal(predict01, target)
ac01
```

```python
ac02 = tf.cast(ac01, dtype="float32")
ac02
```

```python
ac03 = tf.reduce_mean(ac02)
ac03
```

```python
# 정확도를 계산
# 위 과정 합

accuracy = tf.reduce_mean( tf.cast(tf.equal(predict01, target), dtype = tf.float32))
print("="*50)
print("accuracy")
print(accuracy.numpy())
```
