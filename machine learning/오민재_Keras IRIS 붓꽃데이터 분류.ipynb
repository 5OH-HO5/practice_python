```python
from sklearn import datasets
from sklearn.model_selection import train_test_split
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import Adam
import numpy as np
```

```python
data = datasets.load_iris()
data
```

```python
X = data.data
X
```

```python
y = data.target
y
```

```python
y = y.reshape(-1, 1)
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
X_train
```

```python
X_test
```

```python
y_train
```

```python
y_test
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X_train)
```

```python
X_train = scaler.transform(X_train)
X_train
```

```python
from sklearn.preprocessing import OneHotEncoder
enc = OneHotEncoder()
enc.fit(y)
```

```python
y_train_onehot = enc.transform(y_train).toarray()
y_train_onehot
```

```python
y_train_onehot = np.array(y_train_onehot, dtype = "float32")
y_train_onehot
```

```python
y_test_onehot = enc.transform(y_test).toarray()
y_test_onehot
```

```python
y_test_onehot = np.array(y_test_onehot, dtype="float32")
y_test_onehot
```

```python
model = Sequential()
model.add(Dense(16, input_dim = 4, activation = "sigmoid"))
```

```python
model.add(Dense(3, activation = "softmax"))
```

```python
model.summary()
```

```python
model.compile(loss = "categorical_crossentropy", optimizer = Adam(lr = 0.1), metrics = ['acc'])
```

```python
model.fit(X_train, y_train_onehot, epochs = 1000)
```

```python
X_test = scaler.transform(X_test)
X_test
```

```python
predict = model.predict(X_test)
predict
```

```python
predict01 = np.argmax(predict, 1)
predict01
```

```python
target = np.argmax(y_test_onehot, 1)
target
```

```python
predict02 = (predict01 == target)
predict02
```

```python
np.sum(predict02)
```

```python
acc = np.sum(predict02) / len(predict02)
acc
```
