이상값 조회



```python
# count컬럼의 데이터에서 25%에 해당하는 데이터 조회
# np.percentile(train["count"], 25)
q1 = np.percentile(train["count"], 25)
q1
```

```python
# count컬럼의 데이터에서 75%에 해당하는 데이터 조회
# np.percentile(train["count"], 75)
q3 = np.percentile(train["count"], 75)
q3
```

```python
# 75% - 25%를 조회
IQR = q3 - q1
IQR
```

```python
# q1 - (1.5*IQR) 보다 작거나
# q3 + (1.5*IQR) 보다 크거나
# -> 95% 신뢰구간에서 이상치

# 이상치가 아닌 데이터만 리턴:
# q1 - (1.5*IQR) 이상 q3 + (1.5*IQR) 이하

trainWithoutOutliers = train[(train["count"] >= q1 - (1.5*IQR)) & (train["count"] <= (q3 + 1.5*IQR))]

print(train.shape)
print(trainWithoutOutliers.shape)
```

```python
train = trainWithoutOutliers
```

선형회귀

```python
# count를 y에 대입
y = train["count"]
y.head()
```

```python
X = train
# X에서 count컬럼 삭제
X.pop("count")
X.head()
```

```python
# datetime컬럼 삭제
X.pop("datetime")
```

```python
# 캐글의 registered컬럼이 테스트 데이터에 없으므로 삭제
X.pop("registered")
# 전체 날짜 없으므로 삭제
X.pop("day")
# 모든 데이터가 0이므로 삭제
X.pop("minute")
# 모든 데이터가 0이므로 삭제
X.pop("second")
```

```python
from sklearn.model_selection import train_test_split
# 전체 데이터 중 20%를 X_test, y_test에 대입
# 나머지 80%를 X_train, y_train에 대입
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
from sklearn.linear_model import LinearRegression
# 선형회귀 모델을 생성
IModel = LinearRegression()
# X_train, y_train으로 회귀모델을 만들어 a, b1, b2, ... 계산
IModel.fit(X_train, y_train)
```

```python
# X_train으로 자전거 대여 대수를 예측하고 예측한 값을 pred에 대입
pred = IModel.predict(X_train)
```

```python
# X_train, y_train의 square를 계산
IModel.score(X_train, y_train)
```

```python
from sklearn.metrics import mean_squared_error
# 예측값 pred와 진짜값 y_train의 잔차 제곱의 합을 계산
print ("MSE: ", mean_squared_error(pred, y_train))
```

```python
# X_test으로 자전거 대여 대수를 예측하고 예측한 값을 pred에 대입
pred = IModel.predict(X_test)
# 예측값 pred와 진짜값 y_train의 잔차 제곱의 합을 계산
print ("MSE: ", mean_squared_error(pred, y_test))
```

```python
# X_test, y_test의 R square를 계산
IModel.score(X_test, y_test)
```
