```python
# 라이브러리 임포트
# mongodb에 접속 및 명령어 실행
from pymongo import MongoClient
#mongodb에 파일을 저장
from gridfs import GridFS
# mongodb에 저장되어있는 파일을 다운로드
from gridfs import GridFSBucket
```

```python
# mongodb python_test 데이터베이스에 접속 = oracle_xe 같은 역할
db = MongoClient().python_test
```

```python
# python_tset에 파일을 저장할 GridFS 객체 생성
fs = GridFS(db)
```

```python
# url을 요청해서 이미자를 가져오기 쉽게 함
import urllib.request
# 다운로드 받을 이미지 url
url = "http://image.news1.kr/system/photos/2018/11/2/3373952/article.jpg"
```

```python
# 이미지의 확장자 리턴
# [-1] 맨 뒤에서 첫번째
image_type = url.split(".")[-1]
# http://image = [0] / news1 = [1] / kr/system/photos/2018/11/2/3373952/article = [2] / jpg = [3]
image_type
```

```python
# 이미지의 이름 리턴
image_name = url.split("/")[-1]
# [-1] 맨 뒤에서 첫번째
image_name
```

```python
# 이미지의 내용을 읽어서 image에 저장
image = urllib.request.urlopen(url).read()
image
```

```python
# 이미지 파일을 저장할 객체 생성
bucket = GridFSBucket(db)
```

```python
# 이미지 파일을 GridFS에 저장할 객체 grid_in 생성
grid_in = bucket.open_upload_stream(
        "test.jpg", metadata = ["contentType"])
```

```python
# 이미지의 내용을 GridFS에 저장
grid_in.write(image)
```

```python
# 이미지 저장 종료
grid_in.close()
```
