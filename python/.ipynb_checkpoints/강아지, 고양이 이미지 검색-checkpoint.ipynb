```python
import urllib.request
```

```python
# url 가져오기 위해서 해당 이미지 한땀한땀 정성을 다해서 가져와야된다.
# 하지만 naver api 사용 시 검색어 하나로 샤샤샥 가져올 수 있다.
def get_json(keyword, display, start):
    url = "https://openapi.naver.com/v1/search/image?query=" + keyword + "&display" + str(display) + "&start" + str(start)
    
    result = requests.get(url, headers = headers)
    # 가져온 url의 json(link, title 등)이 나옴
    return result.json()
```

```python
def call_and_print(keyword , total_page):
    link_list = []
    for page in range(1,total_page+1):
        json_A = get_json(keyword,20,((page-1)*20)+1)
        for index in json_A['items']:
            link_list.append(index['link'])
    return link_list
```

```python
real_link_list = []
real_link_list = call_and_print("고양이",50)
```

```python
for url in real_link_list:
    try:
        image_type = url.split(".")[-1]
        print(image_type)
        content_type = "image/{}".format(image_type)
        image_name = url.split("/")[-1]
        print(image_name)
        image = urllib.request.urlopen(url).read()
        grid_in = bucket.open_upload_stream(image_name , metadata={"contentType":content_type , "tpye":"cat"})
        grid_in.write(image)
        grid_in.close()
    except:
        print("X")
    
```
