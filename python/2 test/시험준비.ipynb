```python
cctv = pd.read_csv('C:/ai/workspace/data/cctv_in_seoul.csv', encoding='utf-8')
cctv.head()
```

```python
cctv.rename(columns={'기관명':'자치구'}, inplace = True)
cctv.head()
```

```python
data_seoul = pd.merge(seoul, cctv, on='자치구')
data_seoul.head()
```

```python
data_seoul.drop([2], inplace=True)
data_seoul.head()
```

```python
del data_seoul['계']
data_seoul.head()
```

```python
import matplotlib.font_manager as fm
```

```python
f1 = fm.FontProperties(fname='C:/Windows/Fonts/malgun.ttf').get_name()
plt.rc('font', family=f1)

data_seoul['소계'].plot(kind='barh', grid=True, figsize=(10, 5))
plt.show()
```

```python
data_seoul.reset_index()
data_seoul.set_index('자치구', inplace=True)
data_seoul.head()
```

```python
#그래프그리기준비
plt.figure()
#소계데이터를 그래프로 그림
data_seoul['소계'].plot(kind='barh', # 수평방향의 바 그래프
				grid=False, # 그래프의 눈금을 출력하지 않음
                    figsize=(10, 5)) # 그래프의 가로, 세로를 설정
plt.show()
```

```python
data_seoul['소계'].sort_values().plot(kind='barh', #가로 바 그래프를 출력
                                    grid=True, #그래프의 눈금을 출력
                                    figsize=(10,10)) #그래프의 가로, 세로 설정)
plt.show()
```
