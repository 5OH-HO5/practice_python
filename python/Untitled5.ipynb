```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.factorplot('view_num', data = youtube, kind = 'count')
```

```python
youtube['view_num2'] = 0
for index in range(len(youtube)):
    view_num = youtube['view_num'][index]
    if(view_num < 500000):
        youtube['view_num2'][index] = 0
    elif(view_num < 1000000):
        youtube['view_num2'][index] = 1
    else :
        youtube['view_num2'][index] = 2

youtube.head()
```

```python
g = sns.factorplot('view_num2', data=youtube, kind='count')
g.set_xlabels()
```
