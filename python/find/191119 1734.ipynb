```python
from selenium import webdriver
import time
```

```python
browser = webdriver.Chrome("chromedriver.exe")
```

```python
browser.get("http://python.org")
```

```python
top=browser.find_element_by_id('top')
top
```

```python
li_list = top.find_elements_by_tag_name('li')
li_list
```

```python
for li in li_list:
    print(li.text)
    if li.text == 'PyPI':
        print("PyPI 클릭합니다")
        li.click()
        break
```

```python
menu1 = browser.find_elements_by_class_name('pypi-meta')
menu1
```

```python
menu1.click()  #for문으로 해봐야됨
```
