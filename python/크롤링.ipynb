```python
import requests
```

```python
response = requests.get("http://www.weather.go.kr/weather/observation/currentweather.jsp")
```

```python
response.text
```

```python
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(response.text, "html.parser")
```

```python
soup
```

```python
table= soup.find('table',{'class':'table_develop3'})
table
```

```python
table.find_all('tr')
```

```python
data = []
for data1 in table.find_all('tr'):
    #print('data1',data1)
    tds = data1.find_all('td')
    #print('tds',tds)
    for td in tds:
        #print("td",td)
        if td.find('a'):
            #print("td.find('a'):", td.find('a'))
            print("td.find('a').text:", td.find('a').text)
            print('tds[5].text:',tds[5].text)
            point = td.find('a').text
            temperature = tds[5].text
            humidity= tds[10].text
            data.append([point,temperature,humidity])
            #print('tds[10].text:',tds[10].text)
    #print('-'*100)
```
